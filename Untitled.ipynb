```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english')
```

```python
sugg = pd.read_csv('./data/all_sugg_text.csv').set_index('title_id')
```

```python
movies = pd.read_csv('./movies_data/movies.csv')#.sort_values(by = 'year')

movies = movies[['id', 'primary_title', 'year']]

movies = movies[movies['id'].isin(sugg.index)] #https://stackoverflow.com/a/12098586

movies['year'] = movies['year'].astype(int)

movies = movies.sort_values(by = 'year')

title_dict = movies[['id', 'primary_title']].set_index('id').to_dict()['primary_title']
```

```python
sugg
```

```python
sugg_df = pd.DataFrame([title_dict.keys(), title_dict.values()]).T.merge(left_on = 0, right_on = 'title_id', right = sugg)
```

```python
sugg_df.to_csv('./data/suggestion_titles.csv', index = False)
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
X = cv.fit_transform(sugg_df['text'])
```

```python
X_df = pd.DataFrame(X.toarray(), columns = cv.get_feature_names(), index = sugg_df[0])
```
