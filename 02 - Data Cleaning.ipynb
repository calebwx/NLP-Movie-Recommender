```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('./data/movies.tsv', sep = '\t')
movies_2 = pd.read_csv('./data/movies_2.tsv', sep = '\t')
```

```python
ratings = pd.read_csv('./data/ratings.tsv', sep = '\t')
```

```python
movies = movies.merge(right = movies_2, left_on = 'tconst', right_on = 'titleId')

movies = movies.merge(right = ratings, left_on = 'tconst', right_on = 'tconst')

movies = movies[movies['titleType'] == 'movie']
movies = movies[movies['region'] == 'US']

movies_pop = movies[(movies['numVotes'] > 5000)]
```

```python
len(movies_pop.tconst.unique())
```

```python
#Saves a file containing title id (tconst), primary (display) title, and movie year.
#title_reference = movies_pop[['tconst', 'primaryTitle', 'startYear']].drop_duplicates()
#title_reference.to_csv('movies_data/title_reference.csv', index = False)
```

```python
movies_df = pd.DataFrame(columns = ['id', 'titles', 'primary_title', 'year', 'rating', 'num_votes'])

count = 0

for movie_id in movies_pop.tconst.unique():
    mov = movies_pop[movies_pop['tconst'] == movie_id]
    row1 = mov[:1]
    titles = set(list(mov.primaryTitle.values) + list(mov.originalTitle.values) + list(mov.title.values))
    year = row1['startYear'].values[0]
    primary_title = row1['primaryTitle'].values[0]
    rating = row1['averageRating'].values[0]
    num_votes = row1['numVotes'].values[0]
    
    movies_df = movies_df.append({'id' : movie_id,
                                  'titles' : titles,
                                  'year' : year,
                                  'rating' : rating,
                                  'num_votes' : num_votes,
                                  'primary_title' : primary_title},
                                   ignore_index = True)
    
    count += 1
    
    if (count % 100 == 0):
        print(count, end = '...')
```

```python
def set_to_str(strset):
    titles_string = ''
    while True:
        try:
            titles_string += strset.pop() + '%'
        except:
            break
    
    titles_string = titles_string[:-1]
    return titles_string
```

```python
movies_df['titles'] = movies_df['titles'].apply(set_to_str)
```

### Fixing unfortunate titles

After matching titles to suggestions, it became clear that certain movies (whos titles/alternate titles consist of one very common word) were overrepresented in the data. To correct for this, offending titles will be removed, and if necessary, the entire entry will be removed from the data. (in the case where there is no suitable alternative title)

```python
corrections = pd.read_csv('movies_data/title_corrections.csv')
```

```python
for tt_id in corrections['id']:
    movies_df.loc[movies_df['id'] == tt_id, 'titles'] = corrections.loc[corrections['id'] == tt_id, 'titles'].values[0]
```
